```python
Data_Info = pd.DataFrame({"Column": data.columns,"Non-Null Count": data.notnull().sum().values,"Dtype": data.dtypes.values})
display(Data_Info)
```

```python
Null_Value = data.isnull().sum().sort_values(ascending=False)
Null_Value = pd.DataFrame({"Column": Null_Value.index,"Null Values": Null_Value.values})
display(Null_Value)

plt.figure(figsize=(10, 10))
sns.heatmap(data.isnull(), annot=True, cmap='coolwarm')
plt.title("Null Values", fontsize=14, weight='bold')
plt.show()
```

```python
Unique_Value = data.nunique()
Unique_Value = Unique_Value.sort_values(ascending=False)
Unique = pd.DataFrame({"Column": Unique_Value.index,"Unique Values": Unique_Value.values})
display(Unique)

plt.figure(figsize=(10,6))
sns.barplot(data=Unique,x="Unique Values",y="Column",hue="Column",dodge=False,legend=False,palette="viridis")
for index, value in enumerate(Unique["Unique Values"]):
    plt.text(value + 0.2, index, str(value), va='center', fontsize=10)
plt.title("Unique Values", fontsize=14, weight="bold")
plt.xlabel("Number of Unique Values")
plt.ylabel("Columns")
plt.tight_layout()
plt.show()
```

```python
All_People_Interested_Hearing_App = data[data["Interest_in_Hearing_App"].str.contains("Yes", case=False, na=False)]
display(All_People_Interested_Hearing_App)

Features = All_People_Interested_Hearing_App["Desired_App_Features"].dropna().str.split(",").sum()
Features = [f.strip() for f in Features]
Feature_Counts = pd.Series(Features).value_counts()

plt.figure(figsize=(10,5))
sns.lineplot(x=Feature_Counts.index,y=Feature_Counts.values,marker="o",linewidth=2,color="darkred")
plt.title("People Interested in Hearing App", fontsize=14, weight='bold')
plt.xlabel("Features", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

### **Ôº§ÔΩÖÔΩìÔΩÉÔΩíÔΩâÔΩêÔΩîÔΩâÔΩñÔΩÖ Ôº≥ÔΩîÔΩÅÔΩîÔΩâÔΩìÔΩîÔΩâÔΩÉÔΩì**

```python
Early_Hearing_Care_Average = data["Belief_Early_Hearing_Care"].mean()
Average_Data = pd.DataFrame({"Metric": ["Average Belief in Early Hearing Care"],"Value": [round(Early_Hearing_Care_Average, 2)]})
display(Average_Data)

plt.figure(figsize=(5,4))
sns.boxplot(data=data, y="Belief_Early_Hearing_Care", color="lightgreen")
plt.axhline(Early_Hearing_Care_Average, color="red", linestyle="--", label=f"Mean = {Early_Hearing_Care_Average:.2f}")
plt.title("Belief in Early Hearing Care", fontsize=14, weight='bold')
plt.ylabel("Belief Score")
plt.legend()
plt.show()
```

```python
Early_Hearing_Care_Median = data["Belief_Early_Hearing_Care"].median()
Median_Data = pd.DataFrame({"Metric": ["Median Belief in Early Hearing Care"],"Value": [Early_Hearing_Care_Median]})
display(Median_Data)


plt.figure(figsize=(6,4))
sns.boxplot(data=data, y="Belief_Early_Hearing_Care", color="lightcoral")
plt.axhline(Early_Hearing_Care_Median, color="blue", linestyle="--", label=f"Median = {Early_Hearing_Care_Median}")
plt.title("Belief in Early Hearing Care", fontsize=14, weight='bold')
plt.ylabel("Belief Score")
plt.legend()
plt.show()
```

```python
Daily_Headphone_Usage_Mode = data["Daily_Headphone_Use"].mode()[0]
Mode_Data = pd.DataFrame({"Metric": ["Most Common Daily Headphone Usage"],"Value": [Daily_Headphone_Usage_Mode]})
display(Mode_Data)


plt.figure(figsize=(7,4))
sns.countplot(data=data,x="Daily_Headphone_Use",order=data["Daily_Headphone_Use"].value_counts().index,hue="Daily_Headphone_Use",legend=False,palette="pastel")
plt.axhline(data["Daily_Headphone_Use"].value_counts().max(),color="red", linestyle="--",label=f"Mode = {Daily_Headphone_Usage_Mode}")
plt.title("Most Common Daily Headphone Usage", fontsize=14, weight='bold')
plt.xlabel("Daily Headphone Use")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.legend()
plt.show()
```

```python
Early_Hearing_Care_Std = data["Belief_Early_Hearing_Care"].std()
Standard_Data = pd.DataFrame({"Metric": ["Standard Deviation of Belief in Early Hearing Care"],"Value": [Early_Hearing_Care_Std]})
display(Standard_Data)

plt.subplot(1,2,2)
sns.boxplot(x=data["Belief_Early_Hearing_Care"], color="lightgreen")
plt.title("Spread of Belief in Early Hearing Care", fontsize=14, weight='bold')
plt.xlabel("Belief Score")
plt.tight_layout()
plt.show()
```

```python
Belief_Early_Hearing_Care_Variance = data["Belief_Early_Hearing_Care"].var()
Variance_Data = pd.DataFrame({"Metric": ["Variance of Belief in Early Hearing Care"],"Value": [round(data["Belief_Early_Hearing_Care"].var()
, 2)]})
display(Variance_Data)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(data["Belief_Early_Hearing_Care"], kde=True, bins=10, color="skyblue")
plt.axvline(data["Belief_Early_Hearing_Care"].mean(), color="red", linestyle="--", label="Mean")
plt.title("Distribution of Belief in Early Hearing Care", fontsize=14, weight='bold')
plt.xlabel("Belief Score")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Usage_Counts = data["Daily_Headphone_Use"].value_counts()

Minimum_Usage_Headphone = Usage_Counts.idxmin()
Maximum_Usage_Headphone = Usage_Counts.idxmax()

Minimum_and_Maximum_Daily_Headphone_Usage = pd.DataFrame({"Metric": ["Minimum Usage", "Maximum Usage"],"Category": [Minimum_Usage_Headphone, Maximum_Usage_Headphone],"Count": [Usage_Counts.min(), Usage_Counts.max()]})
display(Minimum_and_Maximum_Daily_Headphone_Usage)

plt.figure(figsize=(8,5))
ax = sns.barplot(y=Usage_Counts.index,x=Usage_Counts.values,hue=Usage_Counts.index,dodge=False,palette="pastel",legend=False)
for i, v in enumerate(Usage_Counts.values):
    ax.text(v + 0.5, i, str(v), color="black", va="center")

plt.axhline(y=Usage_Counts.index.get_loc(Minimum_Usage_Headphone), color="green", linestyle="--", label=f"Min: {Minimum_Usage_Headphone}")
plt.axhline(y=Usage_Counts.index.get_loc(Maximum_Usage_Headphone),color="red", linestyle="--", label=f"Max: {Maximum_Usage_Headphone}")

plt.title("Daily Headphone Usage (Min vs Max Highlighted)", fontsize=14, weight='bold')
plt.xlabel("Number of Respondents")
plt.ylabel("Daily Headphone Use")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Minimum_Score = data["Belief_Early_Hearing_Care"].min()
Maximum_Score = data["Belief_Early_Hearing_Care"].max()
Belief_Range  = Maximum_Score - Minimum_Score

Range_Early_Hearing = pd.DataFrame({"Metric": ["Minimum Score", "Maximum Score", "Range"],"Value": [Minimum_Score, Maximum_Score, Belief_Range]})
display(Range_Early_Hearing)

plt.figure(figsize=(6,4))
sns.barplot(x="Metric",y="Value",hue="Metric",data=Range_Early_Hearing,palette="pastel",legend=False)
for i, val in enumerate(Range_Early_Hearing["Value"]):
    plt.text(i, val + 0.0, round(val,2), ha='center', fontsize=12, weight='bold')
plt.title("Range of Belief in Early Hearing Care", fontsize=12, weight="bold")
plt.ylabel("Score")
plt.show()
```

```python
Q1 = data["Belief_Early_Hearing_Care"].quantile(0.25)
Q3 = data["Belief_Early_Hearing_Care"].quantile(0.75)
IQR_value = Q3 - Q1

IQR = pd.DataFrame({"Metric": ["Q1 (25%)", "Q3 (75%)", "IQR"],"Value": [Q1, Q3, IQR_value]})
display(IQR)
plt.figure(figsize=(6,4))
sns.boxplot(x=data["Belief_Early_Hearing_Care"], color="lightcoral", width=0.4)
plt.axvline(Q1, color="blue", linestyle="--", label=f"Q1: {Q1:.2f}")
plt.axvline(Q3, color="green", linestyle="--", label=f"Q3: {Q3:.2f}")
plt.title("Boxplot Showing IQR of Belief in Early Hearing Care", fontsize=14, weight="bold")
plt.xlabel("Belief Score")
plt.legend()
plt.show()
```

```python
Low_Belief_Scores = data[data["Belief_Early_Hearing_Care"] < 3].shape[0]
Total_Respondents = data.shape[0]
Percentage_Low_Belief = (Low_Belief_Scores / Total_Respondents) * 100

Low_Belief_Data = pd.DataFrame({"Metric": ["Low Belief (<3) Count", "Total Respondents", "Percentage Low Belief"],"Value": [Low_Belief_Scores, Total_Respondents, round(Percentage_Low_Belief, 2)]})

display(Low_Belief_Data)

plt.figure(figsize=(6,6))
labels = ["Belief Score < 3", "Belief Score ‚â• 3"]
sizes = [Low_Belief_Scores, Total_Respondents - Low_Belief_Scores]
colors = sns.color_palette("pastel")

plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, wedgeprops={"edgecolor":"blue"})
plt.title("Very Low Belief Scores (<3)", fontsize=14, weight='bold')
plt.show()
```
